```python
import pandas as pd
import requests
import os
import time

# 경로 설정
BASE_DIR = '/content/drive/MyDrive/clickbait_project'
CSV_PATH = f'{BASE_DIR}/dataset.csv'
IMG_DIR  = f'{BASE_DIR}/thumbnails'

os.makedirs(IMG_DIR, exist_ok=True)

# dataset.csv 읽기
df = pd.read_csv(CSV_PATH)
print(f"총 {len(df)}개 데이터 확인")

# 썸네일 다운로드
success = 0
fail = 0

for idx, row in df.iterrows():
    video_id = row['video_id']
    save_path = f"{IMG_DIR}/{video_id}.jpg"

    # 이미 다운받은 거면 스킵 (중간에 끊겨도 이어서 가능)
    if os.path.exists(save_path):
        success += 1
        continue

    # YouTube 썸네일 URL
    url = f"https://img.youtube.com/vi/{video_id}/hqdefault.jpg"

    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            with open(save_path, 'wb') as f:
                f.write(response.content)
            success += 1
        else:
            fail += 1
    except:
        fail += 1

    if idx % 100 == 0:
        print(f"[{idx}/{len(df)}] 완료:{success} 실패:{fail}")

    time.sleep(0.1)

print(f"\n완료! 성공:{success} 실패:{fail}")
```
